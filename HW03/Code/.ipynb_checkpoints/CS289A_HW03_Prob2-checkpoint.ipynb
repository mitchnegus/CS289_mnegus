```python
import math
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def makesquarespace(xmin,xmax,ymin,ymax,delta):
# Create arrays defining the space over which to evaluate density f'n Z
    
    xaxis = np.arange(xmin,xmax,delta)
    yaxis = np.arange(ymin,ymax,delta)
    X,Y = np.meshgrid(xaxis,yaxis)
    Z = np.empty_like(X)
    
    return(X,Y,Z)
```

```python
def calc_density_fn(X,Y,Z,mu,Sigma):
# Given a space defined by X and Y, evaluate the density function Z for a
# bivariate normal distribution with mean mu and covariance matrix Sigma
    n = len(mu)
    detCov = np.linalg.det(Sigma)
    SigInv = np.linalg.inv(Sigma)
    
    for i in range(len(X)):
        for j in range(len(X[i])):
            x = np.array([X[i,j],Y[i,j]])
            Z[i,j] = 1/((2*math.pi)**(n/2)*detCov**(1/2))*math.exp(-0.5*np.dot(np.dot((x-mu),SigInv),(x-mu)))
            
    return Z
```

```python
def contourplots(X,Y,Z):
# Create contour plots with labeled isovalues

    fig = plt.figure(figsize=(10,10))
    cs = plt.contour(X,Y,Z)
    plt.contourf(X,Y,Z)
    cs = plt.clabel(cs, inline=1, fontsize=14, colors='white')
    plt.xlabel("$X_1$")
    plt.ylabel("$X_2$")
    plt.show()
    
```

```python
def SolveAndPlot(xmin,xmax,ymin,ymax,delta,mu,Sigma):
    X1,X2,Z = makesquarespace(xmin,xmax,ymin,ymax,delta)
    Z = calc_density_fn(X1,X2,Z,mu,Sigma)
    contourplots(X1,X2,Z)
    
```

```python
def SolveSubAndPlot(xmin,xmax,ymin,ymax,delta,mu1,Sigma1,mu2,Sigma2):
    X1,X2,Z = makesquarespace(xmin,xmax,ymin,ymax,delta)
    Y1,Y2 = np.empty_like(Z),np.empty_like(Z)
    Y1 = calc_density_fn(X1,X2,Y1,mu1,Sigma1)
    Y2 = calc_density_fn(X1,X2,Y2,mu2,Sigma2)
    Z = Y1-Y2
    contourplots(X1,X2,Z)
    
```

```python
delta = 0.1
```

```python
# (2a)

# Give the mean and covariance matrix (Sigma) of the multivariate distribution
mu = np.array([1,1])
Sigma = np.array([[1,0],[0,2]])
SolveAndPlot(-1,3,-1,3,delta,mu,Sigma)
```

```python
# (2b)

# Give the mean and covariance matrix (Sigma) of the multivariate distribution
mu = np.array([-1,2])
Sigma = np.array([[2,1],[1,3]])
SolveAndPlot(-4,2,-1,5,delta,mu,Sigma)
```

```python
# (2c)

# Give the means and covariance matrices(Sigma1,Sigma2) of the multivariate distribution
mu1,mu2 = np.array([0,2]),np.array([2,0])
Sigma1 = np.array([[2,1],[1,1]])
Sigma2 = np.copy(Sigma1)
SolveSubAndPlot(-2.5,4.5,-2,4,delta,mu1,Sigma1,mu2,Sigma2)
```

```python
# (2d)

# Give the means and covariance matrices(Sigma1,Sigma2) of the multivariate distribution
mu1,mu2 = np.array([0,2]),np.array([2,0])
Sigma1,Sigma2 = np.array([[2,1],[1,1]]),np.array([[2,1],[1,3]])
SolveSubAndPlot(-2.5,4.5,-2,4,delta,mu1,Sigma1,mu2,Sigma2)
```

```python
# (2e)

# Give the means and covariance matrices(Sigma1,Sigma2) of the multivariate distribution
mu1,mu2 = np.array([1,1]),np.array([-1,-1])
Sigma1,Sigma2 = np.array([[2,0],[0,1]]),np.array([[2,1],[1,2]])
SolveSubAndPlot(-3,3,-3,3,delta,mu1,Sigma1,mu2,Sigma2)
```
